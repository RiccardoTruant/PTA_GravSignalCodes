```python
#Library:

import sys
import numpy as np
import math
import random
import gzip
import os
import re

#set the cosmology
from astropy.cosmology import FlatLambdaCDM
from astropy import constants as const
import astropy.units as u
import matplotlib.pyplot as plt
```

```python
#read OUTPUT/GWB_SORT_MBHBs_David_103_Universe_105_Realization_101.dat
#first column freq second column h_c
data = np.loadtxt('OUTPUT/GWB_SORT_MBHBs_David_103_Universe_105_Realization_101.dat')

log10freq = data[:,0]
log10h_c = data[:,1]

#plot

plt.scatter(log10freq, log10h_c,color='red',s=30,marker='v',edgecolors='k',label='GWB from MBHBs')
plt.xlabel(r'$\log10 f_{gw} [Hz]$')
plt.ylabel(r'$\log10 h_c$')
plt.title('GWB Characteristic Strain vs Frequency')
plt.show()
```
